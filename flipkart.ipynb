```python
import urllib
import urllib.request
from bs4 import BeautifulSoup
```

```python
url = "https://www.flipkart.com/televisions/pr?sid=ckf%2Cczl&p%5B%5D=facets.fulfilled_by%255B%255D%3DFlipkart%2BAssured&p%5B%5D=facets.availability%255B%255D%3DExclude%2BOut%2Bof%2BStock&p%5B%5D=facets.serviceability%5B%5D%3Dtrue&p%5B%5D=facets.operating_system%255B%255D%3DAndroid&otracker=clp_banner_1_36.bannerX3.BANNER_television-store_028X6KS22DR4&fm=neo%2Fmerchandising&iid=M_4b2ede7d-ad3d-438a-8e18-3bf2e7547000_36.028X6KS22DR4&ppt=clp&ppn=television-store&ssid=zv9ujcbi8w0000001579009186508"
```

```python
def make_soup(url):
    thepage = urllib.request.urlopen(url)
    source = BeautifulSoup(thepage, 'html.parser')
    return source

```

```python
soup = make_soup(url)
print(soup.title.text)
```
